### **Introducción**
Aproximadamente el 7% de la población colombiana presenta discapacidades auditivas que afectan su calidad de vida, principalmente por la barrera que limita su comunicación por la falta de educación de la población general al no tener necesidad de aprender lengua de señas y dado que esta lengua generalmente es autóctona a cada país [Mapenzi]. Existen diferentes herramientas que buscan brindar un apoyo a esta población, sin embargo, es necesario implementar nuevas estrategias tecnológicas que permitan que la población general integre de manera eficaz a esta población marginada. 

Para el desarrollo del proyecto, se propone una metodología que estará ligada con el avance del curso Técnicas de Inteligencia Artificial de la Universidad Nacional de Colombia, el cual aportará las herramientas necesarias para la implementación de una solución que permita cumplir los objetivos establecidos y el alcance del proyecto. En el siguiente documento se establecen los lineamientos, metodologías y bases necesarias para su adecuado desarrollo. 
 

### **Palabras Clave**
Lengua de señas, IA, OpenCV, LSC, ASL, Visión de máquina.

### **Planteamiento del Problema**
Diseñar un sistema de reconocimiento de lengua de señas colombiana básico, es decir, el alfabeto y gestos sencillos asociados a palabras comunes, con un énfasis en el aprendizaje de dicha lengua. Esto a partir de la necesidad de mejorar las condiciones para esta población y de la falta de desarrollo propio sobre la temática expuesta.

### **Estado del arte**

Encontramos diferentes metódos para la realización del proyecto basandonos en proyectos similares realizados por otros grupos en el pasado:

1.   En la actualidad existen distintas estrategias importantes que se utilizan para este tipo de trabajos de reconocimiento del alfabeto de lengua de señas, muchas utilizan incluso el sistema de reconocimiento del dispositivo Kinect de Microsoft. El reconocimiento principal de este último proceso es el de Bosque aleatorio, con ángulos en las uniones de los dedos de la mano para la identificación de las letras.

![Random Forest](/content/random_forest.png) 

2.   Otro procedimiento de igual importancia que también se realiza es con Redes Convolucionales (CNN) con aumento de la vista múltiple y fusión de inferencia.

![Redes Convolucionales](/content/redes_convolucionales.png)

3. Otra metodología ampliamente utilizada, que no requiere de hardware especializado y que puede ser implementado mediante el uso de cámaras convencionales, es el algoritmo de Support Vector Machine (SVM). En esta metodología  es necesario determinar la posición de los puntos pertenecientes a la figura que se desea clasificar, estos puntos son clasificados según un hiperplano. Al analizar una imagen, se identifica la figura y si los puntos que la componen satisfacen la ecuación del super plano es posible clasificarla según los parámetros establecidos.

![Support Vector Machine](/content/Presentacion/SVM.png)

4.  Otro algoritmo empleado para la clasificación y predicción de figuras es el K-Nearest-Neighbor (KNN). Este algoritmo toma una referencia y la compara con los datos analizados estableciendo un punto de interés que le permite hacer predicciones. Esta metodología se caracteriza por ser supervisada al tener un conjunto de datos base o de entrenamiento y es basada en instancias, ya que el algoritmo no va aprendiendo, sino que por el contrario, memoriza un patrón para poder efectuar las predicciones

![K-nearest](Fig/Presentacion/knn.png)

### **Justificación del proyecto**

En la conversación con una integrante del grupo estudiantil de la Universidad Nacional de Colombia “Mapenzi Inclusiva”, se nos informó sobre cómo la sordera es la discapacidad sensorial más común en Colombia. El grupo sobre el aprendizaje de la Lengua de Señas en Colombia reconoce la sordera como uno de los problemas menos informados en la sociedad pese a la gran población afectada. Como se informa en INSOR a partir del censo nacional del 2018, la población sorda total del país corresponde a 314.320 personas. Viendo todas estas limitaciones en cuanto a la educación de la lengua de señas colombiana en Colombia, se quiere diseñar un sistema que permita el reconocimiento de los gestos más relevantes para ayudar a la educación en la manifestada área. 

### **Objetivo General**
Construir un modelo basado en inteligencia artificial capaz de reconocer expresiones de producción gestual y percepción visual, enfocados al aprendizaje de la lengua de señas colombiana.

#### **Objetivos específicos**
- Obtener una base de datos lo suficientemente grande como para entrenar nuestro algoritmo.
- Investigar soluciones similares a la problemática abordada para afrontar el - problema de la mejor forma y sentar bases para desarrollos futuros.
- Definir los algoritmos de preprocesamiento de imagen que más convengan para el reconocimiento de gestos básicos realizados con las manos.
- Detectar con precisión en tiempo real el abecedario de lengua de señas colombiano.


### **Metodología**

Para lograr tener un ritmo de trabajo apropiado y el desempeño necesario para ejecutar todas las tareas, se proponen dos reuniones a la semana. Las actividades desarrolladas, ya planteadas en el cronograma y las futuras que se planifiquen, están ligadas al propio avance del curso. Es decir, se aplicarán las temáticas de la asignatura en nuestro proyecto.

En cuanto a las herramientas que se utilizarán, nos basaremos en las ya ofrecidas por la universidad junto con las presentadas en clase. La base de datos por ejemplo se recopila en una unidad de Google Drive. Así mismo, los avances del proyecto se documentan en Jupyter Notebook y con la librería de Python OpenCV. 

### **Resultados y productos esperados de la investigación**

Diseñar un algoritmo que nos otorgue un intervalo de confianza lo suficientemente grande para la identificación de expresiones gestuales utilizadas en la lengua de señas. Se plantea por su complejidad limitar el proyecto a solo el abecedario como la primera etapa de este desarrollo. El resultado final es una aplicación capaz de reconocer gestos básicos de la lengua de señas colombiana, enfocada al aprendizaje de esta misma.
